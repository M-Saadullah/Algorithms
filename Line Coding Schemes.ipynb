```python
import matplotlib.pyplot as plt
```

```python
def contains_zeros_and_ones(s):
    for data in s:
        if data!='1' and data!='0':
            return False;
    return True;



def is_valid_choice_of_scheme(user_choice,max_to_choose):
    if(user_choice>=1 and user_choice<=max_to_choose):
        return True
    return False;



def is_valid_choice_to_try_again(choice):
    if(choice!=1 and choice != 0):
        return False;
    return True;



def get_binary_string():
    input_str = input("Enter a binary to generate signal: ") 
    while(not contains_zeros_and_ones(input_str)):
        input_str = input("Please Enter valid string having 0's and 1's only: ")
    return input_str;



def get_binary_string_for_two_B_one_Q():
    input_str = input("Note: Please enter binary of len(binary)%2==0 i.e. even length \nEnter binary: ") 
    while(not contains_zeros_and_ones(input_str) or len(input_str)%2==1):
        input_str = input("Please Enter valid string of even length having 0's and 1's only: ")
    return input_str;
```

```python
def menu():
    
    print("1. NRZ");
    print("2. NRZ-L");
    print("3. NRZ-I");
    print("4. RZ");
    print("5. Manchester");
    print("6. Differential Manchester");
    print("7. AMI");
    print("8. Pseudoternary");
    print("9. 2B1Q");
    choice = int(input("Choose Line coding scheme: "));
    while(not is_valid_choice_of_scheme(choice,9)):
        choice = int(input("Please enter valid no(1-9): "));
        
    return choice;
```

```python
def NRZ(binary):
    # appending extra character only for balancing graph
    binary += "0";
    time_axis = range(len(binary))

    # Polar NRZ-L encoding
    NRZ = [0 if bit == '0' else 1 for bit in binary]
    
    plt.figure(figsize=(10, 3))
    plt.subplot(2, 1, 1)
    plt.axhline(0, color='black',linewidth = 0.5)
    plt.step(time_axis, NRZ, where='post',linewidth = 3)
    plt.title('Unipolar NRZ Encoding')
    plt.xlabel('Time -->')
    plt.ylabel('Voltage Level -->')
    plt.xticks(time_axis, binary)
    plt.ylim(-0.3, 1.3)

    plt.show()
```

```python
def NRZ_L(binary):
    # appending extra character only for balancing graph
    binary += "0";
    
    time_axis = range((len(binary)))
    # Polar NRZ-L encoding from binary string
    polar_NRZ_L = [1 if bit == '0' else -1 for bit in binary]
    
    # Graph
    plt.figure(figsize=(10, 3))
    plt.subplot(2, 1, 1)
    plt.axhline(0, color='black',linewidth = 0.5)
    plt.step(time_axis, polar_NRZ_L, where='post',linewidth = 3)
    plt.title('Polar NRZ-L Encoding')
    plt.xlabel('Time -->')
    plt.ylabel('Voltage Level -->')
    plt.xticks(time_axis, binary)
    plt.ylim(-1.5, 1.5)
    plt.show()
    
# NRZ_L("01001110")
```

```python
def NRZ_I(binary):
    # appending extra character only for balancing graph
    binary+='0';
    
    time_axis = range(len(binary))

    # Polar NRZ-I encoding from binary string
    flag = 1;
    polar_NRZ_I = []
    for bit in binary:
        if(bit == '0'):
            polar_NRZ_I.append(flag)
        else:
            if(flag==1):
                polar_NRZ_I.append(-1)
                flag = -1
            else:
                polar_NRZ_I.append(1)
                flag = 1
    # Graph
    plt.figure(figsize=(10, 3))
    plt.subplot(2, 1, 1)
    plt.axhline(0, color='black',linewidth = 0.5)
    plt.step(time_axis, polar_NRZ_I, where='post',linewidth = 3)
    plt.title('Polar NRZ-I Encoding')
    plt.xlabel('Time -->')
    plt.ylabel('Voltage Level -->')
    plt.xticks(time_axis, binary)
    plt.ylim(-1.5, 1.5)
    plt.show()
```

```python
def RZ(binary):
    time_axis = range(len(binary))
    # binary to RZ
    rz_data = []
    for i in range(len(binary)):
        if binary[i] == '1':
            rz_data.extend([1, 0])  
        else:
            rz_data.extend([-1, 0])  


    plt.figure(figsize=(10, 3))
    # Graph
    plt.subplot(2, 1, 2)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.step(range(len(rz_data)), rz_data, where='post', linewidth=3)
    plt.title('RZ Encoding')
    plt.xlabel('Time -->')
    plt.ylabel('Voltage Level -->')
    plt.xticks(range(0, len(rz_data), 2), binary)
    plt.ylim(-1.5, 1.5)

    plt.tight_layout()
    plt.show()
```

```python
def manchester(binary):
    time_axis = range(len(binary))
    # binary to manchester encoding
    manchester_data = []
    for i in range(len(binary)):
        if binary[i] == '0':
            manchester_data.extend([1, -1])  
        else:
            manchester_data.extend([-1, 1])  


    plt.figure(figsize=(10, 3))
    # Graph
    plt.subplot(2, 1, 2)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.step(range(len(manchester_data)), manchester_data, where='post', linewidth=3)
    plt.title('Manchester Encoding')
    plt.xlabel('Time -->')
    plt.ylabel('Voltage Level -->')
    plt.xticks(range(0, len(manchester_data), 2), binary)
    plt.ylim(-1.5, 1.5)

    plt.tight_layout()
    plt.show()
```

```python
def differential_manchester(binary):
    
    time_axis = range(len(binary))

    # differential_manchester encoding from binary string
    flag = -1;
    differential_manchester_data = []
    for bit in binary:
        if(bit == '0'):
            differential_manchester_data.extend([flag,-flag])
        else:
            if(flag==1):
                differential_manchester_data.extend([-1,1])
                flag = -1
            else:
                differential_manchester_data.extend([1,-1])
                flag = 1
    
    # Graph
    plt.figure(figsize=(10, 3))
    plt.subplot(2, 1, 1)
    plt.axhline(0, color='black',linewidth = 0.5)
    plt.step(range(len(differential_manchester_data)), differential_manchester_data,where = "post",linewidth = 3)
    plt.title('differential_manchester Encoding')
    plt.xlabel('Time -->')
    plt.ylabel('Voltage Level -->')
    plt.xticks(range(0,len(differential_manchester_data),2), binary)
    plt.ylim(-1.5, 1.5)
    plt.show()
```

```python
def pseudoternary(binary):
    # appending extra character only for balancing graph
    binary+='1';
    # converting string data
    time_axis = range(len(binary))

    # AMI encoding from binary string
    flag = 1;
    pseudoternary_data = []
    for bit in binary:
        if(bit == '1'):
            pseudoternary_data.append(0)
        else:
            if(flag==1):
                pseudoternary_data.append(1)
                flag = -1
            else:
                pseudoternary_data.append(-1)
                flag = 1
    plt.figure(figsize=(10, 3))
    plt.subplot(2, 1, 1)
    plt.axhline(0, color='black',linewidth = 0.5)
    plt.step(time_axis, pseudoternary_data, where='post',linewidth = 3)
    plt.title('Pseudoternary Encoding')
    plt.xlabel('Time -->')
    plt.ylabel('Voltage Level -->')
    plt.xticks(time_axis, binary)
    plt.ylim(-1.5, 1.5)
    plt.show()
```

```python
def AMI(binary):
    # appending extra character only for balancing graph
    binary+='0';
    # converting string data
    time_axis = range(len(binary))

    # AMI encoding from binary string
    flag = 1;
    AMI_data = []
    for bit in binary:
        if(bit == '0'):
            AMI_data.append(0)
        else:
            if(flag==1):
                AMI_data.append(1)
                flag = -1
            else:
                AMI_data.append(-1)
                flag = 1
    plt.figure(figsize=(10, 3))
    plt.subplot(2, 1, 1)
    plt.axhline(0, color='black',linewidth = 0.5)
    plt.step(time_axis, AMI_data, where='post',linewidth = 3)
    plt.title('AMI Encoding')
    plt.xlabel('Time -->')
    plt.ylabel('Voltage Level -->')
    plt.xticks(time_axis, binary)
    plt.ylim(-1.5, 1.5)
    plt.show()
```

```python
def two_B_one_Q(binary):
    
    # appending extra characters only for balancing graph
    binary+='11';
    
    time_axis = range(len(binary))
    # binary to two_B_one_Q encoding
    two_B_one_Q_data = []
    for i in range(0,len(binary),2):
        if binary[i] == '0' and binary[i+1]=='0':
            two_B_one_Q_data.append(-3)  
        elif binary[i] == '0' and binary[i+1]=='1':
            two_B_one_Q_data.append(-1)  
        elif binary[i] == '1' and binary[i+1]=='0':
            two_B_one_Q_data.append(3)
        else:
            two_B_one_Q_data.append(1) 


    plt.figure(figsize=(10,6))
    # Graph
    plt.subplot(2, 1, 2)
    plt.axhline(0, color='black', linewidth=0.5)
    plt.step(range(len(two_B_one_Q_data)), two_B_one_Q_data, where='post', linewidth=3)
    plt.title('2B1Q Encoding')
    plt.xlabel('Time -->')
    plt.ylabel('Voltage Level -->')
#     plt.xticks(range(0, len(two_B_one_Q_data), 2), binary)
    plt.ylim(-3.5, 3.5)

    plt.tight_layout()
    plt.show()
```

```python
while(True):
    scheme = menu();
    if(scheme != 9):  # check is for 9th encoding scheme because there may be extra checks in this scheme
        binary = get_binary_string()
        
    
    if(scheme==1):
        # NRZ implimentation
        NRZ(binary);
        
        
    elif(scheme==2):
        # NRZ-L implimentation
        NRZ_L(binary);
        
        
    elif(scheme==3):
        # NRZ-I implimentation
        NRZ_I(binary);
        
        
    elif(scheme==4):
        # RZ implimentation
        RZ(binary);
        
        
    elif(scheme==5):
        # Manchester implimentation
        manchester(binary);
        
        
    elif(scheme==6):
        # Differential manchester implimentation
        differential_manchester(binary);
        
        
    elif(scheme==7):
        # AMI implimentation
        AMI(binary);
        
        
    elif(scheme==8):
        # Pseudoternary implimentation
        pseudoternary(binary);
        
        
    else:
        # 2B1Q implimentation
        binary = get_binary_string_for_two_B_one_Q()
        two_B_one_Q(binary);
    
    choice = int(input("Do you want to try another binary? (1 for yes, 0 for No): "))
    while(not is_valid_choice_to_try_again(choice)):
        choice = int(input("Please Enter only 1/0 for yes/no: "))
    if(choice==0):
        break;
        
print("\nThanks for using system...");
print("Have a nice day...");
```
